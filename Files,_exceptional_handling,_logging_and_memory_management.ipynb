**1 What is the difference between interpreted and compiled languages?**

Compiled Languages: The entire source code is translated into machine code by a compiler before execution (e.g., C, C++). This usually results in faster execution speeds.

Interpreted Languages: Code is executed line-by-line by an interpreter (e.g., Python, JavaScript). This makes debugging easier and the code more portable, though often slower than compiled code.

**2. What is exception handling in Python?**

Exception handling is a process used to manage errors that occur during the execution of a program (runtime errors). It prevents the program from crashing by "catching" the error and allowing the code to respond gracefully.

**3. What is the purpose of the finally block in exception handling?**

The finally block contains code that will always execute, regardless of whether an exception was raised or caught. It is typically used for clean-up actions, such as closing database connections or files.

**4. What is logging in Python?**

Logging is a means of tracking events that happen when some software runs. It allows developers to record messages (errors, warnings, or info) to a file or console to help with debugging and monitoring the application's health.

**5. What is the significance of the __del__ method in Python?**

The __del__ method is known as a destructor. It is called when an object is about to be destroyed or garbage collected, providing an opportunity to release resources like file buffers or network connections.

**6. What is the difference between import and from... import in Python?**

import module: Imports the entire module. You must use the prefix module.name to access its functions or classes.

from module import name: Imports a specific attribute or function directly into the current namespace, allowing you to use name without the module prefix.

**7. How can you handle multiple exceptions in Python?**

You can handle multiple exceptions by:

Using multiple except blocks to catch different error types specifically.

Using a single except block with a tuple of exceptions, e.g., except (ValueError, TypeError):.

**8. What is the purpose of the with statement when handling files?**

The with statement simplifies file handling by acting as a context manager. It ensures that the file is automatically closed as soon as the block of code is finished, even if an exception occurs.

**9. What is the difference between multithreading and multiprocessing?**

Multithreading: Multiple threads run within the same memory space. It is better for I/O-bound tasks (like web scraping).

Multiprocessing: Multiple processes run in separate memory spaces, each with its own Python interpreter. It is better for CPU-bound tasks (like heavy mathematical calculations) because it bypasses the Global Interpreter Lock (GIL).

**10. What are the advantages of using logging in a program?**

Provides a persistent record of application behavior.

Allows for different levels of severity (e.g., DEBUG vs. ERROR).

Easier to debug production issues where a debugger cannot be attached.

Can be configured to send alerts or write to specific destinations (files, emails, etc.).

**11. What is memory management in Python?**

Memory management in Python is the process by which the Python interpreter handles the allocation and deallocation of memory for objects. It involves a private heap where all Python objects and data structures are located, managed automatically by the Python Memory Manager.

**12. What are the basic steps involved in exception handling in Python?**

The basic steps involve:


Try: Wrapping the code that might cause an error in a try block.


Except: Catching the error in an except block to handle it without crashing the program.


Else: (Optional) Executing code if no exceptions were raised.


Finally: (Optional) Executing clean-up code regardless of the outcome.

**13. Why is memory management important in Python?**

Memory management is crucial because it ensures efficient use of resources, prevents memory leaks (where memory is occupied but not used), and ensures that the system does not run out of RAM during the execution of large applications.

**14. What is the role of try and except in exception handling?**

The try block lets you test a block of code for errors, while the except block lets you handle the error if one occurs. This separation allows the program to continue running or fail gracefully instead of terminating abruptly.

**15. How does Python's garbage collection system work?**

Python primarily uses Reference Counting to track how many references exist for an object. When the count reaches zero, the memory is deallocated. It also features a Generational Garbage Collector to detect and fix "reference cycles" (objects referencing each other) that standard counting might miss.

**16. What is the purpose of the else block in exception handling?**

The else block is executed only if the code inside the try block does not raise any exceptions. It is useful for code that should only run if the "protected" code was successful.

**17. What are the common logging levels in Python?**

The common levels, in increasing order of severity, are:

DEBUG: Detailed information for diagnosing problems.

INFO: Confirmation that things are working as expected.

WARNING: An indication that something unexpected happened or indicative of some problem in the near future.

ERROR: Due to a more serious problem, the software has not been able to perform some function.

CRITICAL: A serious error, indicating that the program itself may be unable to continue running.

**18. What is the difference between os.fork() and multiprocessing in Python?**

os.fork(): A Unix-specific system call that creates a child process by copying the parent. It is low-level and not available on Windows.


multiprocessing: A high-level module that works across different platforms (Windows/Linux/macOS) and provides a more Pythonic API for creating and managing processes.

**19. What is the importance of closing a file in Python?**

Closing a file ensures that all data is properly written (flushed) to the disk and releases the system resources/locks associated with that file. Failure to close files can lead to data corruption or hitting the system limit for open file descriptors.

**20. What is the difference between file.read() and file.readline()?**

file.read(): Reads the entire content of the file as a single string.


file.readline(): Reads only one line from the file at a time, moving the pointer to the next line.

**21. What is the logging module in Python used for?**

The logging module is used to record events, errors, and status messages during a program's execution. It is more flexible and professional than print() statements because it can be directed to files, categorized by severity, and formatted with timestamps.

**22. What is the os module in Python used for in file handling?**

The os module is used for interacting with the operating system to perform tasks like renaming files, deleting files, creating directories, and checking if a file path exists.

**23. What are the challenges associated with memory management in Python?**

Challenges include memory fragmentation, where free memory is split into small pieces, and circular references, which can lead to memory not being freed immediately if not handled correctly by the garbage collector.

**24. How do you raise an exception manually in Python?**

You use the raise keyword followed by the exception class or instance.

Example:

    raise ValueError("Only positive numbers are allowed") [cite: 30]

**25. Why is it important to use multithreading in certain applications?**

Multithreading is important for I/O-bound tasks (like downloading files or web scraping) because it allows the program to handle multiple tasks concurrently. While one thread waits for a response from the network, another thread can continue executing, making the application more responsive.

**15. Write a Python program that prints the content of a file and handles the case when the file is empty.**

```python
import os

file_name = "test.txt"

if os.path.exists(file_name):
    if os.stat(file_name).st_size == 0:
        print("The file is empty.")
    else:
        with open(file_name, "r") as file:
            print(file.read())
else:
    print("File not found.")
```

**18. How would you implement a basic logging setup that logs to a file with rotation after 1MB?**

```python
import logging
from logging.handlers import RotatingFileHandler

handler = RotatingFileHandler("app.log", maxBytes=1024*1024, backupCount=3)
logger = logging.getLogger("MyLogger")
logger.addHandler(handler)

logger.warning("This message will trigger rotation if the file exceeds 1MB.")
```

**21. Write a Python program that reads a file and prints the number of occurrences of a specific word.**

```python
target_word = "Python"
count = 0

try:
    with open("notes.txt", "r") as file:
        for line in file:
            words = line.split()
            count += words.count(target_word)
    print(f"The word '{target_word}' appears {count} times.")
except FileNotFoundError:
    print("File not found.")
```
